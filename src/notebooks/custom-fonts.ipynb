# Custom fonts in Matplotlib

This page shows how to import custom fonts in Matplotlib. Today we are going to use two different fronts, coming from two different sources.

* **Hydrophilia Iced**. Available from Floodfonts [here](https://www.floodfonts.com/). [This link](https://www.floodfonts.com/freefont/download/hydrophilia.zip) gives you a zipped folder with the font. In this case we get both `.otf` and `.ttf` types. Make sure you install `.ttf` files only.
* **Special Elite**. Available from Google Fonts [here](https://fonts.google.com/specimen/Special+Elite).

## Steps

1. Download fonts from sources listed above.
2. Install fonts on your system (usually double-click on the `.ttf` file click on Install in the window that opens up). Please note that Matplotlib handles fonts in **True Type Format** (.ttf), so make sure you install fonts ending in `.ttf`
3. Clear matplotlib cache by running the following command on your terminal

```
rm -fr ~/.cache/matplotlib
```

4. If you're using a Jupyter notebook, restart it.
5. Run the following code to tell Matplotlib to look for fonts in your system (it will return a list with font paths)

``` python
from matplotlib import font_manager
font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
```

6. Check our custom fonts are available to Matplotlib with 

```python
font_manager.findfont("Hydrophilia Iced") # Test with the other name too
```

The result should be a path to the font like the following

```
'/home/tomas/.local/share/fonts/HydrophiliaIced-Regular.ttf'
```

### What to do if it fails

If the font is unreachable by Matplotlib you will have a message like

```
findfont: Font family ['Hydrophilia Iced'] not found. Falling back to DejaVu Sans.
```

Try with adding your font into the system library.

On OS X
```
~/Library/Fonts/
```

On Linux

```
/usr/share/fonts/truetype
```

and also just

```
/usr/share/fonts
```

and then repeating steps 2 to 6.

## Plot

We only need to pass the font name to the `fontname` argument and that's it!

````python
import matplotlib.pyplot as plt
````

````python
plt.figure(figsize=(10, 6))
plt.scatter([1, 2, 3, 4], [1, 2, 3, 4], s=500)
plt.text(3, 2, "Hey, I'm a text here!", fontname="Special Elite", fontsize=14)
plt.xlabel("Horizontal Label", fontname="Hydrophilia Iced", fontsize=20)
plt.ylabel("Vertical label too!", fontname="Hydrophilia Iced", fontsize=20)
plt.title("Do you like this title with such a great custom font?", fontsize=28,fontname="Special Elite");
````
