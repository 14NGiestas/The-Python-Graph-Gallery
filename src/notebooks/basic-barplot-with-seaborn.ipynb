Here is an example showing the total bill paid by the customers in different days.
This example produces a basic barplot using the `barplot()` function of [seaborn](http://python-graph-gallery.com/seaborn/) library. The arguments passed to the function are:
* `x,y` : the name of the dataframe columns as inputs
* `data` : dataframe to use for plotting
* `estimator` : function to estimate within each category
* `ci` : size of confidence intervals (if None, error bars will not be drawn)
* `color` : the color of the bars

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
tips = sns.load_dataset("tips")

# Set the figure size
plt.figure(figsize=(14, 8))

# plot a bar chart
ax = sns.barplot(x="total_bill", y="day", data=tips, estimator=sum, ci=None, color='darkblue')
```

It is easy to control bar orders in a seaborn barplot. You can pass the `order` parameter with a list of explicit order to the function.  
I will order the days according to their bill amount in the example below:

```python
# Set the figure size
plt.figure(figsize=(14, 8))

# plot a bar chart with given order of bars ["Fri","Thur","Sun","Sat"]
ax = sns.barplot(x="total_bill", y="day", data=tips, estimator=sum, ci=None, order=["Fri","Thur","Sun","Sat"], color='darkblue')
```
