```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# density plot with a 90 degrees orientation thanks to the 'vertical' parameter
# note: 'vertical' is set to True so that the distribution possible values are on the y-axis, and their respective count/frequency on the x-axis
sns.kdeplot(df['sepal_width'], shade=True, vertical=True, color="skyblue")
plt.show()
```
