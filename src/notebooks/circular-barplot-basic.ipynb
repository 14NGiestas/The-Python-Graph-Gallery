A **circular barplot** is basically a bar plot located on a circle instead of a line. In order to build one, you can use matplotlib library. First, you need to plot polar axis with the `subplot()` function. Then you can add bars to the polar axis as shown in below example. The following parameters are needed:
* `x`: The x coordinates of the bars.
* `height` : The height(s) of the bars.
* `width` : The width(s) of the bars.
* `bottom` : The y coordinate(s) of the bars bases.  


It is also possible to customize bar colors and add text annotations to the bars.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# create data
N = 80
bottom = 2
max_height = 4

theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = max_height*np.random.rand(N)
width = (2*np.pi) / N

labels = []
for i in range(1,81):
    labels.append('Name '+str(i))

# set figure size
plt.figure(figsize=(20,10))

# plot polar axis
ax = plt.subplot(111, polar=True)

# remove grid
plt.axis('off')

# plot bars on the polar axis99+180
bars = ax.bar(theta, radii, width=width, bottom=bottom)

# Use custom colors and opacity
for r, bar in zip(radii, bars):
    bar.set_facecolor("#69b3a2")
    bar.set_alpha(0.8)

# get rotations for text annotation
rotations = np.rad2deg(theta)

# half of the texts should be written reverse
# get the appropriate part of the rotations - their horizontal alignment will be right
rotations[20:60] = rotations[20:60] + 180 % 360 

# add bar labels at the top of each bar
for x, bar, rotation, label in zip(theta[20:60], bars[20:60], rotations[20:60], labels[20:60]):
    ax.text(x,bottom+bar.get_height()+0.3 , label, 
             ha='right', va='center', rotation=rotation, rotation_mode="anchor",) 
for x, bar, rotation, label in zip(theta[0:20], bars[0:20], rotations[0:20], labels[0:20]):
    ax.text(x,bottom+bar.get_height()+0.3 , label, 
             ha='left', va='center', rotation=rotation, rotation_mode="anchor",) 
for x, bar, rotation, label in zip(theta[60:], bars[60:], rotations[60:], labels[60:]):
    ax.text(x,bottom+bar.get_height()+0.3 , label, 
             ha='left', va='center', rotation=rotation, rotation_mode="anchor",) 

# show the graph
plt.show()
```
