If you have different groups and subgroups, you can show their values in a **grouped barplot**. It is possible to show subgroups by pasing the `hue` argument to the `barplot()` function of seaborn. You should pass the column name which have the subgroups in your data. In this example, the subgorups are smokers and non-smokers:

```python
# import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# load dataset
tips = sns.load_dataset("tips")

# Set the figure size
plt.figure(figsize=(14, 8))

# plot a bar chart
ax = sns.barplot(x="day", y="total_bill", hue="smoker", data=tips, estimator=sum, ci=None)
```
